```python
import pandas as pd

df = pd.read_csv('/content/newPriceDataSet.csv')
```

```python
new_df= df.dropna()
print(df)
print(new_df)
```

```python
new_df = df.fillna(60)
print(new_df)
```

```python
x = df["weight"].mean()
new_df=df.fillna(x)
print(new_df)
```

```python
x = df["weight"].median()
new_df=df.fillna(x)
print(new_df)
```

```python
x = df["weight"].mode()[0]
new_df=df.fillna(x)
print(new_df)
```
